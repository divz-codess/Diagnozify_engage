```python
app = Flask(__name__)
run_with_ngrok(app)
# model.summary()
import os
def getpredictions(filename):
    model = keras.models.load_model("my_model.h5")
  
    img_height = 120
    img_width  = 120  
    
    image_url = tf.keras.utils.get_file('Court', origin=filename )
    img1 = tf.keras.preprocessing.image.load_img(image_url, target_size=(img_height, img_width))  
    print(image_url)
    
    os.remove(image_url) # Remove the cached file
    # img1 = image.load_img(filename,target_size=(img_height, img_width))
    # img1 = image.load_img(filename,target_size=(img_height, img_width))
    
 
    Y = image.img_to_array(img1)
    
    X = np.expand_dims(Y,axis=0)
    
    val = model.predict(X)

    print(val)
    val = list(val[0])
    
    
    print(val)
    class_of_object = ['DiGeorge_Syndrome', 'Down_syndrome', 'Noonan_Syndrome', 'Normal_people', 'William_buren_syndrome']

    result = {}
    for i in range(0,len(val)):
        result[class_of_object[i]] = str(val[i])

    
    print(result)
    
    temp = max(val)
    
    max_index = val.index(temp)
    
    print(class_of_object)

    return result
    # print(f"Class={class_of_object[max_index]}, Probability={temp}")


@app.route('/')
def home():
    return jsonify({"Success":"Welcome to Diagnozifyzz API Service !"})

@app.route('/predict',methods=['POST'])
def getprediction():
  try:
    data = request.json
    print(data)
    imageUrl = data['image_url']
    print(imageUrl)
    res = getpredictions(imageUrl)
    return jsonify(
        
        {
            "success":True,
            "msg":"successfull prediction !",
            "data":res
        }
    )   
  except Exception as e:
    print(e)
    return jsonify({"success":False,"msg":"wrong formated image !","data":None},)

app.run()

```
